```python
filename = filenames[68]
```

```python
#Get data and visualize raw data

# get current working directory and go back 2 folder to main working directory
cwd = os.path.normpath(os.getcwd() + os.sep + os.pardir + os.sep + os.pardir)   
load_data_path = '/data/d01_data_raw/d01_load_data_all_raw/'
path = cwd + load_data_path + filename

# Get Data in Dataframe
df = pd.read_csv(path, usecols=["Datum", "Zeit", "Wert"], encoding= 'unicode_escape')
df.Datum = pd.to_datetime(df['Datum'] + ' ' + df['Zeit'])
df = df.set_index('Datum', drop=True).drop_duplicates(keep='first')
df.drop('Zeit', axis=1, inplace=True)
df = df.loc[(df.index >= '2019-01-01') & (df.index < '2022-01-01')]

# Visualize Data
df.Wert.plot(title="KIT load profile from 2019-2021 [kW]",
             figsize=(15,3), 
             xlabel="Time", 
             ylabel="Load [kW]",
             kind="line"
            )
```

```python
#Preprocess data and visualize

#Describe 
display(df.describe().transpose())
#Display Outliers
display("Outliers: " + str(df.loc[(df.Wert - df.Wert.rolling(3).median().abs()) > 38, "Wert"].count()) + 
       " Negative Outliers: " +  str(df.loc[df.Wert < 0, "Wert"].count()) + 
       " Missing values: " +  str(df.loc[df.Wert == 0, "Wert"].count()) + 
        " NaN values: " +  str(df.loc[df.Wert == np.nan, "Wert"].count())
       )

#Cleaning
#No negative values: Replace negative values,with median of last 3 values
df.loc[df.Wert < 0, "Wert"] = df.Wert.rolling(3).median().abs()

#No outliers: Replace values which are 38 higher or lower from the median of a length-3 rolling window with the median
df.loc[(df.Wert > (df.Wert.rolling(3).std().abs())), "Wert"] = df.Wert.rolling(3).median().abs()

#Reduce missing values: Fill 3 missing values by polynomial interpolation
df.loc[df.Wert == 0, "Wert"] = np.nan
df.Wert.interpolate(method='linear', inplace=True, limit=3)

#Set missing values to 0
df.loc[df.Wert.isnull(), "Wert"] = 0


#Plot Result df
_ = df.Wert.plot(title="IPE load profile from 2007-2008 [kW]",
             figsize=(15,3), 
             xlabel="Time", 
             ylabel="Load [kW]",
             kind="line"
            )
display("Missing values: " +  str(df.loc[df.Wert == 0, "Wert"].count()))
```
