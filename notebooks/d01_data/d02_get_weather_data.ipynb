# Task
This notebook downloads weather data from meteostat and stores it in a csv file

Source weather data: <https://meteostat.net/de/place/de/karlsruhe?s=10727&t=2023-01-17/2023-01-24>

Features: 
- temp [°C]: Air Temperature (Temperatur)
- dwpt [°C]: Dew Point (Taupunkt)
- rhum [%]: Relative Humidity (Relative Luftfeuchtigkeit)
- prcp [mm]: Total Precipitation (Gesamtniederschlag)
- snow [?]: Snow Depth (Schneehöhe)
- wdir [Degrees]: Wind (From) Direction (Windrichtung)
- wspd [km/h]: Average Wind Speed (Windgeschwindigkeit)
- wpgt [km/h]: Wind Peak Gust (Spitzenböe)
- pres [hPa]: Sea-Level Air Pressure (Luftdruck)
- tsun [Min]: Total Sunshine Duration (Sonnenscheindauer)
- coco [Integer]: Weather Condition Code - 1 Clear, 27 Strom (Wetterzustands Code)

! The weather station "karlsruhe" -> 10727 is missing some values, therefor data from the next nearest weather station "Rheinstetten"-> 10731 are used in that case

# Import

```python
from datetime import datetime, timedelta, date
from meteostat import  Daily, Hourly
import pandas as pd 
```

# Get Data

```python
# Get data from Karlsruhe (ID: 10727)

# Set time period
start = datetime(2018, 12, 31, 23, 59)
end = datetime(2021, 12, 31, 23, 59)
# Get hourly data
df = Hourly('10727', start, end)
df = df.fetch()
#Describe 
display(df.describe().transpose(),df.head())
```

# Visualize Data

```python
# Visualize Temperature
df.temp.plot(title="Karlsruhe Temperatur",
             figsize=(15,3), 
             xlabel="Time", 
             ylabel="°C",
             kind="line"
            )
```

```python
# Visualize Average Wind Speed
df.wspd.plot(title="Karlsruhe Average Wind Speed",
             figsize=(15,3), 
             xlabel="Time", 
             ylabel="km/h",
             kind="line"
            )
```

```python
# Visualize Wind Peak Gust
df.wpgt.plot(title="Karlsruhe Wind Peak Gust",
             figsize=(15,3), 
             xlabel="Time", 
             ylabel="km/h",
             kind="line"
            )
```
